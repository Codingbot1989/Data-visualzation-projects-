```python
covid=data.drop(['Sno','Time','ConfirmedIndianNational','ConfirmedForeignNational'],axis=1)
```

```python
covid.head()
```

```python
covid['Date']=pd.to_datetime(covid['Date'],format='%Y-%m-%d')
```

```python
covid.head()
```

```python
#Active cases

covid['Active_cases']=covid['Confirmed']- (covid['Cured'] + covid['Deaths'])
covid.tail()
```

```python
statewise=pd.pivot_table(covid,values=['Confirmed','Deaths','Cured'],index='State/UnionTerritory', aggfunc=max)
```

```python
statewise['Recovery rate']=statewise['Cured']*100/statewise['Confirmed']
```

```python
statewise['Mortality']=statewise['Deaths']*100/statewise['Confirmed']
```

```python
statewise= statewise.sort_values(by='Confirmed',ascending=False)
```

```python
statewise.style.background_gradient(cmap='cubehelix')
```

```python
#Top 10 active cases states

Top_10_most_active_states=covid.groupby(by= 'State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
```

```python
fig=plt.figure(figsize=(16,9))
```

```python
plt.title('Top_10_most_active_cases_in_India',size=25)
```

```python
ax=sns.barplot(data=Top_10_most_active_states.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')
```

```python


Top_10_most_active_states=covid.groupby(by= 'State/UnionTerritory').max()[['Active_cases','Date']].sort_values(by=['Active_cases'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title('Top_10_most_active_cases_in_India',size=25)
ax=sns.barplot(data=Top_10_most_active_states.iloc[:10],y='Active_cases',x='State/UnionTerritory',linewidth=2,edgecolor='red')
plt.xlabel('States')
plt.ylabel('Total Active Cases')
plt.show()
```

```python
# Top states with highest deaths

top_10_deaths=covid.groupby(by= 'State/UnionTerritory').max()[['Deaths','Date']].sort_values(by=['Deaths'],ascending=False).reset_index()
fig=plt.figure(figsize=(16,9))
plt.title('Top_10_deaths',size=25)
ax=sns.barplot(data=top_10_deaths.iloc[:12],y='Deaths',x='State/UnionTerritory',linewidth=2,edgecolor='black')
plt.xlabel('States')
plt.ylabel('Deaths')
plt.show()
```

```python
#growth trends

fig=plt.figure(figsize=(12,6))

ax=sns.lineplot(data=covid[covid['State/UnionTerritory'].isin(['Maharashtra','Karnataka','Kerala','Tamil Nadu','Uttar Pradesh'])],
                                                             x = 'Date',y ='Active_cases',hue ='State/UnionTerritory')

ax.set_title('top 5 affected states in India',size=16)
```

```python
#remove rows where state=india
data=vaccine[vaccine.State!='India']
data
```

```python
data.rename(columns={'Total Individuals Vaccinated':'Total'},inplace=True)
```

```python
data.head()
```

```python
#states with most number of vaccinated
max_vac=data.groupby('State')['Total'].sum().to_frame('Total')
max_vac=max_vac.sort_values('Total',ascending=False)[:5]
max_vac
```

```python
fig=plt.figure(figsize=(10,9))
plt.title('Top 5 Vaccinated states in India',size=25)
x=sns.barplot(data=max_vac.iloc[:10],y=max_vac.Total,x=max_vac.index,linewidth=2,edgecolor='red')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```

```python
#states with min number of vaccinated
min_vac=data.groupby('State')['Total'].sum().to_frame('Total')
min_vac=max_vac.sort_values('Total',ascending=True)[:5]
min_vac
```

```python
fig=plt.figure(figsize=(10,9))
plt.title('Top 5 least Vaccinated states in India',size=20)
x=sns.barplot(data=min_vac.iloc[:10],y=min_vac.Total,x=min_vac.index,linewidth=2,edgecolor='red')
plt.xlabel('States')
plt.ylabel('Vaccination')
plt.show()
```
