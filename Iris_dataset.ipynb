Visualization on iris dataset

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as snr
```

Loading iris dataset from sklearn library

```python
iris=snr.load_dataset('iris')
```

```python
iris.head()
```

```python
iris.tail()
```

```python
#line graph
ax=iris.plot(title='Iris datset',figsize=(15,8),)
ax.set_xlabel('x axis')#total number of rows in iris dataset
ax.set_ylabel('y axis')#highest number of numerical variables
```

```python
#drop the species column
data=iris.drop(['species'],axis=1)#axis=1 means entire column will be deleted from that data
data.head()
```

```python
#bar graph
data.plot(kind='bar')
```

```python
#not understanding because it is overlapped and not very helpful
```

```python
data.head()
```

#instead of showing the entire bar graph we can slice

```python
data.iloc[0].plot(kind='bar') #iloc help in slicing a particular row from entire row
#visualization the 0th row in the from of bar graph
```

```python
data.iloc[1].plot(kind='bar') #1st row
```

```python
data.iloc[2].plot(kind='bar') #2nd row
```

```python
data.iloc[3].plot(kind='bar') #3rd row
```

```python
data.iloc[4].plot(kind='bar') #4th row
```

```python
#for multiple rows                   #visualization for 0th row,1st row,2nd row,3rd row
data.iloc[0:4].plot(kind='bar')
```

```python
data.iloc[0:4].plot(kind='barh') #h stands for horizontal
```

```python
data.plot(kind='hist')
```

```python
data.diff().plot(kind='hist',bins=200) #bins means regulationg width of the graph. Thickness will increase
```

```python
#This looks more like a guassian curve and the whole iris dataset is balanced in nature, but if you drive towards each column it will be imbalance, but overall it ressamble a guassian curve.
```

```python
data['sepal_length'].diff().plot(kind='hist',bins=200)
```

```python
data['petal_length'].diff().plot(kind='hist',bins=200)
```

```python
data['petal_width'].diff().plot(kind='hist',bins=200)
```

```python
data['sepal_width'].diff().plot(kind='hist',bins=200)
```

```python
data.iloc[0].plot(kind='pie')#pie chart for 0th row
```

```python
data.iloc[0].plot.pie(figsize=(10,10),fontsize=15,autopct='%.0f')
```

```python
data.iloc[0:3].plot.pie(subplots=True,figsize=(10,10),fontsize=15,autopct='%.0f')
```
