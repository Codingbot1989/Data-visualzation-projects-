#PAIRPLOT

```python
snr.pairplot(data,x_vars=['TV','Radio','Newspaper'],y_vars=['Sales'],size=10)
```

#The above graph is comparision between sales and tv, sales and radio and sales and newspaper.so we can see that the datapoint in Tv is cross somewhere around 300 and newspaper its somewhere aroud 150, but in Radio its just 50 if we observe the x-axis. So we can conclude that through all these the major contribution of sales is into TV, Newspaper then Radio

```python
#HEATMAP- Will show the corelation between the features
plt.figure(figsize=(5,5))
snr.heatmap(data.corr(),annot=True)
```

#In the above graph we can ee the heatmap where we can see the color intensity as when we compare TV with TV the correlation is 1 and it is highly correlated. where as TV with newspaper it cannot be correlated as TV is electronic device and Newspaper is visual and tangible. Similarly with Radio and TV, Radio and newspaper.
