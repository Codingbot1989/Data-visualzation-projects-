Relational plot using tips dataset

```python
import numpy as np
import pandas as pd
import matplotlib as matplt
import seaborn as snr
```

```python
data=snr.load_dataset('tips')
```

```python
data.head()
```

```python
snr.relplot(x='total_bill',y='tip',data=data)
```

```python
snr.relplot(x='total_bill',y='tip',data=data,hue='smoker')
```

```python
snr.relplot(x='total_bill',y='tip',data=data,hue='smoker',style='time') #comparaing between total_bll,tip,smoker and time
```

```python
snr.relplot(x='total_bill',y='tip',data=data,hue='sex',style='time',size='size')#comparaing between total_bll,tip,smoker,time and size
```

```python
snr.relplot(x='total_bill',y='tip',data=data,hue='smoker',col='time',size='size',style='sex')#comparaing between total_bll,tip,smoker,time,size and sex
```

```python
snr.relplot(x='total_bill',y='tip',data=data,hue='sex',col='day',size='size',style='time') # comapraing total_bill and tip between different days
```

```python
snr.catplot(x='total_bill',y='tip',data=data)
```

```python
snr.catplot(x='total_bill',y='day',data=data)#day if you want to see in one graph
```
